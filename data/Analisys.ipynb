# Carga de datos

```python
#Lectura de datos
df = pd.read_csv("laptop-price.csv")

#Conversion de Indias rupias a dolares
tasa_cambio = 0.0118025
df["Price"] = df["Price"] * tasa_cambio
df.to_csv("Price", index=False)

#Columnas irrelevantes
df.drop(columns=["Hybrid", "index"], inplace = True)

df
```

# Descripcion de las variables
**Index**: El el indice que realciona la posicion del respectivo dato de una fila.

**Company**: Es la compañia de la laptop

**TypeName**: Es tipo de lapptop

**Ram**: El monsto de ram dado en (GB)

**Weight**: El peso de la laptop en (Kg)

**Price**: El precio de la laptop en USD (inicialmente en IND Rupee)

**TouchScreen**: Indica si la laptop cuanta con tactil en la pantalla (Se muestra para el analisis en binario SI/NO)


**Ips**: Indica si la tablep cuenta con en su pantalla con la tecnologia de "Crtistales liquidos alineados en paralelo". Siglas en ingles (In-Plane Switching). Es un tipo de tecnologia de visualizacion que proporciona una mejor experiencia visual al usuario (Se muestra para el analisis en binario SI/NO)

**ppi**: Indica los pixeles por pulgada con los que cuenta la pantalla de la laptop, entre mas tenga se tendra una mayor calidad de visualizacion de imagen en el monitor.

**Cpu brand:** Es la marca de la unidad de procesamiento. La cpu es la unidad de procesmamiento la cual le permite a la laptop realizar calculos mas complejos y procesar mejor los datos

**HDD**: Unidad de disco duro, es la parte que se encarga del almacenamiento, a diferencia de una SSD, esta cuenta con partes mecanicas, escribe los datos y los almacena magenticamente en discos circulares

**SSD**: Disco de estado solido. Es la unidad de almacenamiento. Cuenta con una mejor tecnologia respecto a la HDD, se caracteriz en su ausencia de partes moviles

**Hybrid**: Sistema de alamcenamiento que combina la unidad de disco duro (HDD) y el disco de estado solido (SSD) (Se muestra para el analisis en binario SI/NO)

**Flash_Storage**: Es una tecnologia de almacenamiento. Hoy en dia no se diferencia del disco de estado solido (SSD), Estado dado en (GB)

**Gpu Brand**: Es es la marca de la unidad de procesamiento de graficos 

**os**: Sistema operativo con el que cuenta la laptop
