```python
import requests
import pandas as pd
```

```python
#Key de NewsAPI
API_KEY = "9553be1f920348c8811deb79d3fc97c3"

# URL de la API para obtener noticias sobre tecnología
url = f"https://newsapi.org/v2/top-headlines?category=technology&language=en&apiKey={API_KEY}"

# Hacer la solicitud
response = requests.get(url)
data = response.json()

#Extraer artículos de las noticias
articles = data.get("articles", [])

# DataFrame con las noticias
df = pd.DataFrame(articles)[["source", "author", "title", "description", "url", "publishedAt"]]

# Ver las primeras filas
print(df.head())
```
